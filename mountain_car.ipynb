```python
import matplotlib.pyplot as plt
import gym
from agent import *
plt.style.use('seaborn')
```

```python
env = gym.make('MountainCar-v0')
```

```python
def plot_agent_rewards(agent):
    fig, ax = plt.subplots(1,figsize=(10,8))
    ax.plot(agent.rewards)
    ax.set_title("Agent Reward Progression across Episodes", fontsize = 16)
    ax.set_ylabel('Reward', fontsize = 14)
    ax.set_xlabel('Episodes', fontsize = 14)
    plt.show()
    training_time = agent.get_train_time()
    print('Training Time: {:.2f}s'.format(training_time))
    print("Number of episodes: ", n_episodes)
    print("Batch Size: ", batch_size)
    print("Discount Factor: ", gamma)
    print("Learning Rate: ", lr)
    print("Epsilong Decay: ", decay)
```

```python
agent1=AgentMountainCar(env=env)

n_episodes=250
batch_size=128
gamma = 0.99
lr = 0.001
decay = 0.95

agent1.init_hyperparameters(n_episodes, batch_size, gamma, lr, decay)
agent1.train()
```

```python
plot_agent_rewards(agent1)
```

The agent finds a way to get on top of the hill but it isn't very stable for a while and near the end we see that the agent seems to have reach the top of the mountain  for most episodes.

```python
agent2=AgentMountainCar(env=env)

n_episodes=400
batch_size=256
gamma = 0.95
lr = 0.001
decay = 0.99

agent2.init_hyperparameters(n_episodes, batch_size, gamma, lr, decay)
agent2.train()
```

```python
plot_agent_rewards(agent2)
```

Here we changed the discount factor and it seems a lot less stable than the previous one. It looks like there is progressively better rewards until around episode 200 where it finds the top of the hill but doesn't seem to be very consistent later on. This could be due to Discount Factor being too low and not allowing proper update of value function. The number of episodes was also increased to give more time to learn. The time is much longer than it should be and that is because the computer spent some time sleeping between the episodes.

```python
agent3=AgentMountainCar(env=env)

n_episodes=400
batch_size=128
gamma = 0.995
lr = 0.0001
decay = 0.99

agent3.init_hyperparameters(n_episodes, batch_size, gamma, lr, decay)
agent3.train()
```

```python
plot_agent_rewards(agent3)
```

Here we see that the discount Factor is too high and perhaps the learning rate too low. The agent figured out how to get to the top of the hill but didn't do it enough consecutively to be stable and then went back to swinging back and forth without reach the top of the hill. The time is much longer than it should be and that is because the computer spent some time sleeping between the episodes.
